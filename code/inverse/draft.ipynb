```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import ticker
```

```python
path = "nn/loss"
df = pd.read_csv(path, delim_whitespace=True)
print(df.columns)
n = df['N'].to_numpy()
alpha = df['alpha'].to_numpy()
loss = df['loss'].to_numpy()
time = df['time'].to_numpy()
pred_alpha = df['pred_alpha'].to_numpy()
```

```python
N = n.reshape(5, 5)
Alpha = alpha.reshape(5, 5)
Loss = loss.reshape(5, 5)
Time = time.reshape(5, 5)
Pred_alpha = pred_alpha.reshape(5, 5)
Alpha_error = np.abs(Alpha-Pred_alpha)/Alpha
fig = plt.figure()
ax = fig.add_subplot(111)
"""
img = ax.imshow(Loss, interpolation="nearest",
           extent=[alpha.min(), alpha.max(), N.min(), N.max()], aspect='auto')
"""
img = ax.contourf(N, Alpha, Alpha_error, locator=ticker.LogLocator())
ax.set_aspect((np.max(N) - np.min(N))/(np.max(alpha) - np.min(alpha)))
cbar = fig.colorbar(img)
```
